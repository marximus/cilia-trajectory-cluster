```python
tracks = np.load('sample_data/1001-8_roi_0_tracks.npy')
tracks = tracks[np.random.choice(len(tracks), 25, True)]
```

```python
# compute distances between trajectories
dist = distance.pdist(tracks['trajectory'])
squaredist = squareform(dist)
```

```python
# remove mean and scale to unit variance
# std_scale = StandardScaler().fit(squaredist)
# squaredist_std = std_scale.transform(squaredist)
# print('mean after standardization: {}'.format(squaredist_std.mean(axis=0)))
# plot distance matrix
fig, ax = plt.subplots(1, 1, tight_layout=True)
sns.heatmap(squaredist, ax=ax, square=True, annot=True, annot_kws={'fontsize':6}, fmt='.1f')
# sns.heatmap(squaredist_std, ax=ax[1], square=True, annot=True, annot_kws={'fontsize':6}, fmt='.1f')
```
