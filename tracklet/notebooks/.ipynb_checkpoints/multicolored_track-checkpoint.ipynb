```python
%matplotlib notebook
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import numpy as np
# plt.style.use('seaborn-notebook')
```

```python
# read tracklet data
tracks = np.load('sample_data/1001-8_roi_0_tracks.npy')

# tracks['frame_num'] only contains the frame each tracklet ends on. Expand
# it to contain the frame for each point in the tracklet.
n_lines, n_pts, _ = tracks['coords'].shape
lines = tracks['coords']
t = np.mgrid[:n_lines, :n_pts][1] - n_pts + tracks['frame_num'][:, None] + 1
```

```python
print((lines.shape, t.shape))
segments = []
color = []
for i in range(len(lines)):
    points = lines[i].reshape(-1, 1, 2)
    segs = np.concatenate([points[:-1], points[1:]], axis=1)
    segments.extend(list(segs))
    color.extend(list(t[i]))
segments = np.array(segments)
color = np.array(color)
print((segments.shape, color.shape))
    

# create continuous norm to map points to colors
fig, ax = plt.subplots()
norm = plt.Normalize(t.min(), t.max())
lc = LineCollection(segments, cmap='viridis', norm=norm, linewidth=0.3)
lc.set_array(color)
col = ax.add_collection(lc)
ax.autoscale(True)
fig.colorbar(col, ax=ax)
plt.show()
```
