```python
# %matplotlib notebook
import sys
import os
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
import numpy as np
plt.style.use('seaborn-notebook')
```

```python
# read tracklet data
tracks = np.load('sample_data/1001-8_roi_0_tracks.npy')
```

```python
# compute directional histogram
sys.path.append(os.path.abspath('..'))
from tracklet import features
dhist, bins = features.directional_histogram(tracks['coords'], n_bins=10)
```

```python
# show directional histogram for a couple of tracklets
n_tracklets = 3
fig, ax = plt.subplots(n_tracklets, 2, figsize=(8, 8))

for i in range(n_tracklets):
    ax[i, 0].plot(*tracks['coords'][i].T, '.-')
    ax[i, 0].margins(0.1, 0.2)
    ax[i, 0].set_aspect('equal')
    for t, (x, y) in enumerate(tracks['coords'][i]):
        ax[i, 0].annotate('t={}'.format(t), np.array((x, y))-0.05, fontsize=8)
    ax[i, 1].bar((bins[:-1] + bins[1:]) / 2, dhist[i], align='center', width=0.7*(bins[1] - bins[0]))

plt.show()
```
